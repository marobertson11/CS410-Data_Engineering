```python
import numpy as np
import pandas as pd
import math
NaN = float("nan")
df = pd.read_csv('./hwy26Data.csv')
dflen = len(df["Crash ID"]) 
def exiAss(): # Every crash has a crash ID
    rowWrong=[]
    rowNum=0
    for item in df["Crash ID"]:
        if item == NaN:
            print("Assertion 1 wrong on row #", rowNum, sep='')
            rowWrong.append(rowNum)
        rowNum+=1
    if not rowWrong:
        print("Assertion 1: 'Every crash has a crash ID' is correct")

def limAss(): #Every crash ID has a record type value between 1-3
    rowNum = 0
    wrongRow=[]
    for item in df["Record Type"]:
        if (item < 1 or item > 3):
            print("Assertion 2 Wrong on Row #", rowNum, sep='')
            wrongRow.append(rowNum)
        rowNum+=1
    if not wrongRow:
        print("Assertion 2: 'Every crash ID has a record type value between 1-3' is Correct")

def intraAss(): #Every record type 1 entry has a serial number
    rowWrong3=[]
    for i in range(0,dflen):
        if(df['Record Type'][i] == 1):
            if (df['Serial #'][i] == NaN):
                print("Assertion 3 Wrong on Row #", i+1, sep='')
                rowWrong3.append(i)
    if not rowWrong3:
        print("Assertion 3: 'Every record type 1 entry has a serial number' is Correct")
    
def interAssA(): #Each crash has 1 record type of record type 1
    crashIDWrong4 = []
    crashIdder=df.iloc[0,0]
    for i in range(0,dflen):
        if (crashIdder != df.iloc[i,0]):
            if (df.iloc[i,1] != 1):
                print("Assertion 4A Wrong on Row #", i +1, sep='')
                crashIDWrong4.append(crashIdder)
            crashIdder = df.iloc[i,0]
    if not crashIDWrong4:
        print("Assertion 4A: 'Each crash has 1 record type of record type 1' is Correct")

def interAssB(): #Every crash has at least 1 entry of record type 2
    crashIdder=df.iloc[0,0]
    stillGood = False
    crashidWrong=[]
    for i in range(0,dflen):
        if (crashIdder != df.iloc[i,0]):
            if (stillGood == False):
                print("Assertion 4B Wrong on Crash ID#", crashIdder, sep='')
                crashidWrong.append(crashIdder)
            stillGood = False
            crashIdder = df.iloc[i,0]
        if (df.iloc[i,1] == 2):
            stillGood = True
        if (i == dflen-1):
            if (stillGood == False):
                print("Assertion 4B Wrong on Crash ID#", crashIdder, sep='')
                crashidWrong.append(crashIdder)
    if not crashidWrong:
        print("Assertion 4B: 'Every crash has at least 1 entry of record type 2' is Correct")

def sumAssA(): #On average, crashes involve 2 or more vehicles
    crashIdder=df.iloc[0,0]
    outliers1=[]
    numbs=[]
    count=0
    total=0
    for i in range(0,dflen):
        if (crashIdder != df.iloc[i,0]):
            numbs.append(count)
            if (count < 2):
                outliers1.append(crashIdder)
            count = 0
            crashIdder = df.iloc[i,0]
        if (df.iloc[i,1] == 2):
            count += 1
        if (i == dflen-1):
            numbs.append(count)
            if (count < 2):
                outliers1.append(crashIdder)
            count = 0
    for item in numbs:
        total+=item
    avg = total / len(numbs)
    avg = round(avg,2)
    if avg >= 2:
        print("Assertion 5A: 'On average, crashes involve 2 or more vehicles' is Correct")
        print("The outliers for this assertion are the following IDs...")
        for it in range(len(outliers1)):
            if (it==len(outliers1)-1):
                print(outliers1[it])
            else:
                print(outliers1[it], end=', ')
    else:
        print("Assertion 5A: 'On average, crahses involve 2 or more vehicles' is false, the average is actually", avg)
            
def sumAssB(): #Most commonly, crashes involve 2 participants (meaning the mode for the number of participants in each crash is 2)
    crashIdder=df.iloc[0,0]
    outliers2=[]
    numbs=[]
    count=0
    total=0
    for i in range(0,dflen):
        if (crashIdder != df.iloc[i,0]):
            numbs.append(count)
            if (count < 2):
                outliers2.append(crashIdder)
            count = 0
            crashIdder = df.iloc[i,0]
        if (df.iloc[i,1] == 3):
            count += 1
        if (i == dflen-1):
            numbs.append(count)
            if (count < 2):
                outliers2.append(crashIdder)
            count = 0
    for item in numbs:
        total+=item
    avg = total / len(numbs)
    if avg >= 2:
        print("Assertion 5B: 'Most commonly, crashes involve 2 participants' is Correct")
    print("There are", len(outliers2), "outliers for this assertion, and have the following IDs...")
    for it in range(len(outliers2)):
        if (it==len(outliers2)-1):
            print(outliers2[it])
        else:
            print(outliers2[it], end=', ')
        
def statAssA(): #0.	Crashes are less likely to occur on a Monday.
    days=[0,0,0,0,0,0,0]
    crashIdder=0
    total=0
    flag = True
    #days[df.iloc[12,1]-1] += 1
    for i in range(0,dflen):
        if (crashIdder != df.iloc[i,0]):
            indx = int(df.iloc[i,11]-1)
            days[indx] += 1
            crashIdder = df.iloc[i,0]
    for i in range(0,len(days)):
            if(i == 1): continue # This is monday
            else:
                   if days[i] <= days[1]:
                       print("Day code ", i+1, " disapproves assertion 6A with ", days[i], " crashes on this day of the week", sep='')
                       flag = False
    if flag == True:
        print("Assertion 6A: 'Crashes are less likely to occur on a Monday (Code 2)' is correct")
    else:
        print("Monday (Code 2) has a total of ", days[1], " crashes", sep='')

def statAssB(): #The majority of crashes occur when the weather conditions are a level 2 or higher
    weather=[0,0]
    crashIdder=0
    total=0
    flag = True
    for i in range(0,dflen):
        if (crashIdder != df.iloc[i,0]):
            crashIdder = df.iloc[i,0]
            if(int(df.iloc[i,55]) == 1):
                weather[0] += 1
            else:
                weather[1] += 1
    if (weather[0] < weather[1]):
        print("Assertion 6B: 'The majority of crashes occur when the weather conditions are a level 2 or higher'")
    else:
        print("Assertion 6B is incorrect, crashes occured with weather condition 1 was ", weather[0], ",\n while there was a total of ", weather[1], " crashes recorded when the level was higher than 1", sep='')

exiAss()
limAss()
intraAss()
interAssA()
interAssB()
sumAssA()
sumAssB()
statAssA()
statAssB()
```
