```python
#Change the uniplaces.txt to uniplaces.csv, not 100% accurate since the txt isn't 100% right
import re
f = open("./uniplaces.txt", 'r')
lines = f.readlines()
f.close()
state=""
city=""
school=""
file = open("./uniplaces.csv", 'w')
file.write("State")
file.write(", ")
file.write("City")
file.write(", ")
file.write("School")
file.write("\n")
for line in lines:
    statematch = re.search("^\w*\[edit]", line)
    if (statematch):
        state=""
        for item in line:
            if item == "[":
              break
            state+=item
    else:
        paren=0
        schoolFlag=False
        for item in line:
            if item =="(":
                schoolFlag=True
                continue
            if item == ")" or item=="[": #occasionally there is not end parenthesis
                file.write(state)
                file.write(", ")
                file.write(city)
                file.write(", ")
                file.write(school)
                file.write("\n")
                schoolFlag=False
                city=""
                school=""
                break
            if schoolFlag:
                if item == ",": #multiple colleges in same city
                    file.write(state)
                    file.write(", ")
                    file.write(city)
                    file.write(", ")
                    file.write(school)
                    file.write("\n")
                    school=""
                    continue
                school+=item
            else:
                if(item == ","):
                    city+="-"
                    continue
                city+=item
file.close()
```
